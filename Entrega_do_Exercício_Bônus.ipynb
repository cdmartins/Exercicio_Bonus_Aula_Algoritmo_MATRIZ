```python
matriz = None
while True:
    print()
    print("***EXERCÍCIO COM MATRIZES***")
    print(28 * "*")
    print("Selecione a opção desejada: ")
    print(28 * "*")
    print("1 - Preencher uma matriz com números inteiros")
    print("2 - Imprimir a matriz de números inteiros")
    print("3 - Imprimir a soma dos elementos de cada linha da matriz")
    print("4 - Imprimir a soma dos elementos de cada coluna da matriz")
    print("5 - Imprimir a soma dos elementos da diagonal principal")
    print("6 - Imprimir a matriz transposta")
    print("7 - Imprimir quantidade de elementos pares e ímpares")
    print("8 - Imprimir a soma de todos os elementos da matriz")
    print("9 - Imprimir o maior elemento da matriz e onde se encontra")
    print("0 - Encerrar o Sistema")
    print()
    opcao = int(input(" "))

    # OPÇÃO 1: preencher a matriz deixando usuario escolher quantas linhas e quantas colunas, informando a posicão de cada vez que ele for preencher a sua matriz.
    if opcao == 1:
        linhas = int(input("Digite a quantidade de LINHAS(s) da matriz: "))
        colunas = int(input("Digite a quantidade de COLUNAS(s) da matriz: "))
        matriz = []
        for i in range(linhas):
            linha = []
            for j in range(colunas):
                valor = int(input(f"Digite o valor para a posição [{i}] [{j}]: "))
                linha.append(valor)
            matriz.append(linha)
    # OPÇÃO 2: Imprimir a matriz que ele criou em formato de matriz
    elif opcao == 2:
        if matriz is not None:
            for linha in matriz:
                print(linha)
        else:
            print(
                "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
            )
    # OPÇÃO 3: Fazer a soma de cada linha da matriz e mostrar o resultado.
    elif opcao == 3:
        if matriz is not None:
            somalinha = 0
            for i in range(linhas):
                for j in range(colunas):
                    somalinha = somalinha + matriz[i][j]
                print(f"A soma da linha {i}, é: {somalinha}")
                somalinha = 0
        else:
            print(
                "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
            )
    # OPÇÃO 4: Imprimir a soma dos elementos de cada coluna da matriz
    elif opcao == 4:
        if matriz is not None:
            somacoluna = 0
            for j in range(colunas):
                for i in range(linhas):
                    somacoluna = (somacoluna + matriz[i][j])
                print(f"A soma da coluna {j}, é: {somacoluna}")
                somacoluna = 0
        else:
            print(
                "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
            )

    # OPÇÃO 5: Imprimir a somados elementos da diagonal principal
    elif opcao == 5:
        if matriz is not None:
            somadiagonal = 0
            for i in range(linhas):
                for j in range(colunas):
                    if i == j:
                        somadiagonal += matriz[i][j]
            print(f"A soma dos elementos da diagonal principal é: {somadiagonal}")

        else:
            print(
                "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
            )

    # OPÇÃO 6: Imprimir a matriz transposta onde a linha vira coluna
    elif opcao == 6:
        if matriz is not None:
            matriz_transposta = []
            for j in range(colunas):
                linha_transposta = []
                for i in range(linhas):
                    linha_transposta.append(matriz[i][j])
                matriz_transposta.append(linha_transposta)
            print("Matriz transposta: ")
            for linha in matriz_transposta:
                print(linha)
        else:
            print(
                "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
            )

    # OPÇÃO 7: Imprimir quantidade de elementos pares e impares
    elif opcao == 7:
        if matriz is not None:
            qtd_pares = qtd_impares = 0
            for i in range(linhas):
                for j in range(colunas):
                    if (matriz[i][j] %2 == 0):
                        qtd_pares += 1
                    else: qtd_impares += 1
            print(f'A quantidade de números PARES presentes na sua matriz é {qtd_pares}')
            print(f'A quantidade de números ÍMPARES na sua matriz é: {qtd_impares}')
        else:
            print(
                "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
            )
    #OPÇÃO 8: Imprimir a soma de todos os elementos da matriz
    elif opcao == 8:
            if matriz is not None:
                somageral = 0
                for i in range(linhas):
                    for j in range(colunas):
                        somageral = (somageral + matriz[i][j])
                print(f"A soma de todos os elementos da sua matriz é: {somageral}")

            else:
                print(
                    "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
                )
    #OPÇÃO 9: Imprimir o maior elemento da matriz e onde se encontra
    elif opcao == 9:
            if matriz is not None:
                maior=linha_maior=coluna_maior= 0
                for i in range(linhas):
                    for j in range(colunas):
                        if (matriz[i][j]>maior):
                            maior=matriz[i][j]
                            linha_maior = i
                            coluna_maior = j
                print(f'O maior elemento é: {maior} e encontra-se na linha {linha_maior} e na coluna {coluna_maior}')


            else:
                print(
                    "A matriz não foi preenchida ainda. Por favor, escolha a opção 1 para preencher a matriz primeiro."
                )
    elif opcao == 0:
        break
```
